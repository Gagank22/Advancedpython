```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Basic Approach of Pie Chart
```

```python
Pie1=np.array([35,45,90,10])
plt.pie(Pie1)
plt.show()
```

```python
mylabels = ['A','B','C','D']    #Giving label names
plt.pie(Pie1,labels=mylabels)
plt.show()
```

```python
# Controlling the angles in Pie Chart
```

```python
plt.pie(Pie1,labels=mylabels,startangle=45)
plt.show()
```

```python
myexplode=[0.2,0,0,0]
plt.pie(Pie1,labels=mylabels,explode=myexplode)
plt.show()
```

```python
plt.pie(Pie1,labels=mylabels,explode=myexplode,shadow=True) #Shadowing the Pie Chart
plt.show()
```

```python
plt.pie(Pie1,labels=mylabels,explode=myexplode,shadow=True)
plt.legend()
plt.show()
```
